```python
print("Welcome to Hash Book Store!!") #greeting Massage

stock = {
    "math":{"qty": 5, "price":250},
    "science":{"qty": 8, "price":230},
    "english":{"qty": 7, "price":290},
    "hindi":{"qty": 9, "price":270},
    "social":{"qty": 3, "price":260},
}

def buy_book(book_name): #Function for buy Book
    if book_name in stock:
        buy = input(f"Yes {book_name} is available. It's cost around {stock[book_name]["price"]} You want to buy? (yes/no)").lower()
        if buy == "yes":
            stock[book_name]["qty"] -=1
            print(f"You buy {book_name} so your total is {stock[book_name]["price"]}")
        else:
            print("No worries may be next time!!")
    else:
        print(f"I'm sorry the {book_name} book requested is not available")
    return

def add_stock(book1): #Function for add Book
    if book1 in stock:
        print("Oh! we already have this book.")
        exit_book = input(f"Do you need some more book? (yes/no)")
        qty1 = int(input("How many pieces you want to add: "))
        if exit_book =="yes":
            stock[book1]["qty"] += qty1
            print(f"{book1} Book add in stock thank you!")
        else:
            print("See you Later")
    else:
        print("Oh! it's new Book. I need it")
        qty1 = int(input("How many pieces do you need."))
        price1 = int(input("What's the price of this Book."))
        stock[book1] = {'qty': qty1, 'price': price1}
        print(f"New Book {book1} now added in stock Thank you!")
    return
    
def retun_book(book_name1): #Function for return Book
    qty2 = int(input("How many pieces you want to return"))
    stock[book_name1]["qty"] += qty2
    print(f"You return the {book_name1} book. Thank you!")
    return

def remove_stock(books): #Function for remove Book
    qty3 = int(input("How many pieces you want to remove"))
    stock[books]["qty"] -= qty3
    if stock[books]["qty"] ==0:
        del stock[books]
    print(f"{books} is now removed from stock")
    return

while True: # Loop for user input
    user = input("How can I help you!!: ").lower()
    if "buy" in user:
        book = input("Which Book you want to buy: ")
        buy_book(book)
    elif "add" in user:
        book2 = input("Which Book you get here: ")
        add_stock(book2)
    elif "return" in user:
        book3 = input("Which Book you want to return: ")
        retun_book(book3)
    elif "remove" in user:
        book4 = input("Which Book you want to remove from stock: ")
        remove_stock(book4)
    elif "exit" in user:
        print("Thank you for coming. See you later!!")
        break
    elif "stock" in user:
        for book_name, qty in stock.items():
            quantity = stock[book_name]["qty"]
            print(f"Book: {book_name}, Quantity: {quantity}")
    else:
        print("Invalid please try again")
```
